```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import io
```

```python
img = io.imread('Lena.png', as_gray=True)
f = np.array(img) * 255
```

```python
plt.imshow(f)
plt.show()
```

```python
M, N = f.shape
```

```python
def dft_row(M):
    x = np.arange(0, M).reshape(M,1)
    u = np.arange(0, M).reshape(1,M)
    return np.exp(-2j*np.pi*x.dot(u)/M)
```

```python
def dft_column(N):
    y = np.arange(0, N).reshape(N,1)
    v = np.arange(0, N).reshape(1,N)
    return np.exp(-2j*np.pi*y.dot(v)/N)
```

```python
def F():
    return dft_row(M).dot(f).dot(dft_column(N))
```
