```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
from scipy.fftpack import dct, idct
```

### Implementar em Python as seguintes funções:
a) **def dct2(.)**, que retorna a DCT em 2D

```python
def dct2(img):
    return dct(dct(img.T).T)
```

b) **def idct2(.)**, que retorna a inversa

```python
def idct2(img):
    return idct(idct(img.T).T)
```

### Carregar "lena.jpg" e em seguida:
a) Calcular e mostrar a DCT



```python
img = imageio.imread('Lena.png')[:,:,0]
plt.imshow(img, cmap='gray')
plt.show()
```

```python
result = idct2(dct2(img))
plt.imshow(result, cmap='gray')
plt.show()
```

b) Calcular a DCT para cada bloco 8x8

```python
def bdct2(img):
    h, w = img.shape
    result = np.zeros((h, w))

    for i in range(0, h, 8):
        for j in range(0, w, 8):
            result[i:(i+8),j:(j+8)] = dct2(img[i:(i+8),j:(j+8)])
    return result
```

```python
result = bdct2(img)
plt.figure(figsize=(100,100))
plt.imshow(result, cmap='gray')
plt.show()
```

```python
def ibct2(img):
    h, w = img.shape
    result = np.zeros((h, w))

    for i in range(0, h, 8):
        for j in range(0, w, 8):
            result[i:(i+8),j:(j+8)] = idct2(img[i:(i+8),j:(j+8)])
    return result
```

```python
plt.imshow(ibct2(result), cmap='gray')
plt.show()
```

c) Se $X=$**dct2(.)**, agora filtre $X$, mantendo somente componentes onde $|X|\geq \tau.max|X|$, $\tau=0.01$, faça a IDCT, e mostre o resultado.

```python
def f(img, T):
    x = dct2(img)
    max = abs(x).max()
    result = (abs(x) >= T*max)*x
    return result
```

```python
r = idct2(f(img, 0.01))
plt.imshow(r, cmap='gray')
plt.show()
```
