```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Sinal unidimensional:

```python
def x(t):
    return 0.75*np.sin(3*t) + 0.5*np.sin(7*t)
```

```python
t = np.arange(0.0, 12.5, 0.01)

plt.plot(t, x(t))
plt.xlabel('time(s)')
plt.ylabel('y label here')
plt.title('title here')
plt.grid(True)
plt.show()
```

### Sinal bidimensional:

```python
def f(x,y):
    return 1.5*np.cos(2*x)*np.cos(7*y) + 0.75*np.cos(5*x)*np.sin(3*x) - 1.3*np.sin(9*x)*np.cos(15*y) + 1.1*np.sin(13*x)*np.sin(11*y)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x, y = np.meshgrid(np.linspace(0, 1, 1000), np.linspace(0, 1, 1000))

ax.plot_surface(x, y, f(x,y))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
