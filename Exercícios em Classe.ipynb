```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x,y):
    return 128 + 128*np.sin(2*np.pi*(50*x+70*y))
```

# 3D Projection

```python
n_samples = [60, 100, 300, 1000]

for n in n_samples:
    x = np.linspace(0, 1, n)
    y = np.linspace(0, 1, n)

    fig = plt.figure()
    ax = fig.gca(projection='3d')

    x, y = np.meshgrid(x, y)

    ax.plot_surface(x, y, f(x,y), antialiased=False, cmap=plt.cm.gray)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_title('Samples: {}'.format(n))
    
    plt.show()
```

# Density

```python
for n in n_samples:
    x = np.linspace(0, 1, n)
    y = np.linspace(0, 1, n)

    fig = plt.figure()

    x, y = np.meshgrid(x, y)
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Samples: {}'.format(n))
    plt.imshow(f(x,y), cmap='gray')
    
    plt.show()
```
