```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft2, fftshift
from scipy.ndimage import rotate
from scipy.ndimage.filters import gaussian_filter
```

```python
side = 512
shft = 150
```

```python
black = np.full((side,side), 0)
for i in range(shft, side-shft):
    for j in range(shft, side-shft):
        black[i][j]=255
plt.figure(figsize=(10,10))
plt.imshow(black, cmap='gray')
plt.show()
```

```python
a = fft2(black)
ampl = np.log(1+abs(a))
plt.figure(figsize=(10,10))
plt.imshow(fftshift(ampl), cmap='gray')
plt.show()
```

```python
black45 = rotate(black, 45)
plt.figure(figsize=(10,10))
plt.imshow(black45, cmap='gray')
plt.show()
```

```python
b = fft2(black45)
ampl45 = np.log(1+abs(b))
plt.figure(figsize=(10,10))
plt.imshow(fftshift(ampl45), cmap='gray')
plt.show()
```

```python
blurred = gaussian_filter(black45, sigma=3)
plt.figure(figsize=(10,10))
plt.imshow(blurred, cmap='gray')
plt.show()
```

```python
c = fft2(blurred)
amplblur = np.log(1+abs(c))
plt.figure(figsize=(10,10))
plt.imshow(fftshift(amplblur), cmap='gray')
plt.show()
```
